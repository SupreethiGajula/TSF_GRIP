# we shall see the comparision between different attributes


we are going to see the comparisions between:

shipmode,sales

segment,sales

region,sales

and finally

sales and profit

```python
import matplotlib.pyplot as plt

# Calculate the average sales for each shipment type
average_sales_by_shipment = df.groupby('Ship Mode')['Sales'].sum()

# Create the bar plot
plt.bar(average_sales_by_shipment.index, average_sales_by_shipment)
plt.xlabel('Shipment Type')
plt.ylabel('Total Sales')
plt.title('Comparison of Sales by Shipment Type')
plt.show()
#from this we can see that the total sales are more in the standard type
```

```python
import matplotlib.pyplot as plt

# Calculate the average sales for each shipment type
average_sales_by_shipment = df.groupby('Ship Mode')['Sales'].mean()

# Create the bar plot
plt.bar(average_sales_by_shipment.index, average_sales_by_shipment)
plt.xlabel('Shipment Type')
plt.ylabel('Average Sales')
plt.title('Comparison of Sales by Shipment Type')
plt.show()
#where as the average sales are almost same in all the types
```

```python
region_encoded = pd.get_dummies(df['Region'])
region_counts = region_encoded.sum() 
```

```python
colors = plt.get_cmap('Blues_r')
color_list = colors(range(len(region_counts)))  # Generate a list of colors
#fig,ax = plt.subplots()
plt.pie(region_counts,colors = color_list,radius = 1,labels=region_counts.index)
plt.title('Region Distribution')
#from the pie chart the inference is there are more of residents in the west and east region,
#so the
#sales may go up there if there is more marketing
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
#we are seeing the total sales for each region using the groupby function
region_sales
```

```python
plt.figure(figsize=(8, 6))
region_sales.plot(kind = 'bar',color = 'green')
plt.title("sales in each region")
plt.xlabel('region')
plt.ylabel('sales')
plt.show()
```

```python
# Calculate the average sales for each shipment type
average_sales_by_shipment = df.groupby('Category')['Sales'].mean()

# Create the bar plot
plt.bar(average_sales_by_shipment.index, average_sales_by_shipment)
plt.xlabel('category')
plt.ylabel('Average Sales')
plt.title('Comparison of Sales by category')
plt.show()
#from this we can say that the average sales are very high in the technology category
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of profit and a continuous attribute
plt.scatter(df['Sales'], df['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Profit vs Sales')
plt.show()

```

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['lime','green'],figsize=(15,8))
plt.title('Profit/Loss & Sales Across States')
plt.xlabel('States')
plt.ylabel('Profit/Loss & Sales')
plt.show()
```

# let us see the comparision between other attributes and the profit

shipmode,segment,city,state,region,category,subcategory vs profit

```python
import seaborn as sns

# Create a bar plot of profit by a categorical attribute
sns.barplot(x='Region', y='Profit', data=df)
plt.xlabel('Region')
plt.ylabel('Profit')
plt.title('Profit by Region')
plt.show()
#high profit in west
```

```python

# Create a bar plot of profit by a categorical attribute
sns.barplot(x='Category', y='Profit', data=df)
plt.xlabel('Category')
plt.ylabel('Profit')
plt.title('Profit by Category')
plt.show()
#high profit in technology
```

```python
sns.barplot(x='Ship Mode', y='Profit', data=df)
plt.xlabel('shipment mode')
plt.ylabel('Profit')
plt.title('Profit by shipment')
plt.show()
#first class has the highest profit
```

```python
p_sub = df.groupby('Sub-Category')[['Profit']].sum().sort_values(by='Profit',ascending=False)
p_sub[:].plot.bar(color=['blue','red'],figsize=(15,8))
plt.title('Profit/Loss by Sub-category')
plt.xlabel('Subcategory')
plt.ylabel('Profit/Loss')
plt.show()
#tables have loss and copiers have the highest profit
```

```python
p_state = df.groupby('State')[['Profit']].sum().sort_values(by='Profit',ascending=False)
p_state[:].plot.bar(color=['red','blue'],figsize=(15,8))
plt.title('Profit/Loss  Across States')
plt.xlabel('States')
plt.ylabel('Profit/Loss')
plt.show()
```

```python
p_city = df.groupby('City')[['Profit']].sum().sort_values(by='Profit',ascending=False)
p_state[:].plot.bar(color=['green','yellow'],figsize=(15,8))
plt.title('Profit/Loss  Across cities')
plt.xlabel('cities')
plt.ylabel('Profit/Loss')
plt.show()
```

```python
sns.barplot(x = 'Segment',y = 'Profit',data = df)
plt.title('segment and profit')
plt.xlabel('Segment')
plt.ylabel('profit')
#home office has the highest profit
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_columns].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
#By selecting only the numeric columns using the select_dtypes function, 
#we can calculate the correlation matrix without encountering the ValueError. 
#The resulting heatmap will show the correlations between the numeric columns in your dataset.
```
