```python
df['city'] = df['city'].fillna(df['city'].mode())
df['winner'] = df['winner'].fillna(0)
df['player_of_match'] = df['player_of_match'].fillna(0)
df['umpire2'] = df['umpire2'].fillna(0)
df['umpire3'] = df['umpire3'].fillna(0)
df['umpire1'] = df['umpire1'].fillna(0)

```

```python
df.shape
#there are 756 rows and 18 columns
```

```python
#who won the matches many times
win = df['winner'].value_counts()
win
```

```python
#plotting the above observations
win.plot(kind = 'pie',autopct='%1.1f%%')
#the highest percentage is shown by mumbai indians i.e., they won the matches many times
```

```python
pof = df['player_of_match'].value_counts()
pof
```

```python
#CH GAyle is the player of the match many times
pof = pof = df['player_of_match'].value_counts()[0:10]
#plot the graph for the top 10 players of the match
pof.plot(kind = 'bar')
```

```python
toss_win = df['toss_winner'].value_counts()
toss_win
```

```python
toss_win.plot(kind = 'pie',autopct = "%1.1f%%")
#again it is mumbai indians who won the toss many times
```

```python
win_run = df.groupby('player_of_match')['win_by_runs'].sum()
win_run
```

```python
win_run.plot(kind = 'bar',figsize = (30,8))
#CH Gayle is the player who won by most runs
```

```python
win_run_team = df.groupby('winner')['win_by_runs'].sum()
win_run_team
```

```python
win_run_team.plot(kind = 'pie',autopct = '%.1f%%')
#it's mumbai indians again followed by chennai super kings
```

```python
win_wi = df.groupby('player_of_match')['win_by_wickets'].sum()
win_wi
```

```python
win_wi.plot(kind  = 'bar',figsize = (30,8),color = 'green')
#it's G Gambhir who won by securing highest number of wickets
```

```python
win_wi_team = df.groupby('winner')['win_by_wickets'].sum()
win_wi_team
```

```python
win_wi_team.plot(kind = 'bar')
```
